```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification, make_regression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from scipy.stats import mode
from scipy.spatial.distance import mahalanobis , minkowski , cityblock , cosine , hamming
```

```python
X, y = make_classification(n_samples=2000, n_features=2, n_informative=2, n_redundant=0,
                           n_clusters_per_class=1, n_classes=2, class_sep=1.2, random_state=15)

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=19)

plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train, s=50, cmap='jet', alpha=0.5)
# plt.scatter(x_test[:, 0], x_test[:, 1], c=y_test, s=50, cmap='jet', marker='s')
```

### *'euclidean' , 'mahalanobis' , 'menhatan' , 'minco' , 'cosine'*

```python
class KNNClassification:
    
    def __init__(self, k, metric='euclidean'):
        self.k = k
        self.metric = metric
        
    def fit(self, X, y):
        self.X = X
        self.y = y
        self.inv_cov = np.linalg.inv(np.cov(X.T))
            
    def predict(self, x , p=2):
        y_hats = []
        for xi in x:
            
            if self.metric == 'euclidean':
                distances = np.linalg.norm(self.X - xi, axis=1)
            elif self.metric == 'mahalanobis':
                distances = np.array([mahalanobis(xi, Xi, self.inv_cov) for Xi in self.X])
            elif self.metric == 'menhatan':
                distances = np.array([cityblock(xi, Xi) for Xi in self.X])
            elif self.metric == 'minco':
                distances = np.array([minkowski(xi, Xi , p) for Xi in self.X])
            elif self.metric == 'cosine':
                distances = np.array([cosine(xi, Xi) for Xi in self.X])
            elif self.metric == 'hamming':
                distances = np.array([hamming(xi, Xi) for Xi in self.X])
                
            indices = np.argsort(distances)[:self.k]
            k_nearest_labels = list(map(lambda i: self.y[i], indices)) #[self.y[i] for i in indices]
            y_hat = mode(k_nearest_labels).mode
            y_hats.append(y_hat)   
        return np.array(y_hats)
```

```python
fig , ax = plt.subplots(2 , 3 , figsize=(16 , 10))
ax_flatt = ax.flatten()

metrics = ['euclidean' , 'mahalanobis' , 'menhatan' , 'minco' , 'cosine']

for i, metric in enumerate(metrics):
    model = KNNClassification(10, metric=metric)
    model.fit(x_train, y_train)
    y_hat = model.predict(x_test)
    y_hat = y_hat.ravel()
    print(f'Number of True Prediction : {np.sum(y_hat == y_test)}')
    dmin, dmax = 0.5, 0.5
    xmin, xmax = X.min(0)-dmin, X.max(0)+dmax
    x1 = np.linspace(xmin[0], xmax[0], 50)
    x2 = np.linspace(xmin[1], xmax[1], 50)
    xm1, xm2 = np.meshgrid(x1, x2)
    xf1, xf2 = xm1.flatten(), xm2.flatten()
    Xf = np.stack((xf1, xf2), axis=1)

    yf = model.predict(Xf)
    ym = yf.reshape(xm1.shape)

    ax_flatt[i].contourf(xm1, xm2, ym, cmap='jet');
    ax_flatt[i].scatter(X[:, 0], X[:, 1], c=y, s=10, cmap='jet', edgecolors='w');
    ax_flatt[i].set_title(f'metric is :{metric}')
```
